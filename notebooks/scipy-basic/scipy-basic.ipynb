```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
n = 10
p = 0.5

# Range of possible successes
k = np.arange(0, n+1)
# Binomial probability mass function
pmf = binom.pmf(k, n, p)

# Plot
# plt.figure(figsize=(8, 6))
plt.vlines(x=k, ymin=0, ymax=pmf, colors='b', lw=5)
plt.plot(k, pmf, 'bo', ms=8)
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability')
plt.title(f'Binomial Distribution PMF (n={n}, p={p})')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters
nA = 2000
pA = 0.2
nB = 2000
pB = 0.25

# Range of possible number of clicks
k = np.arange(0, 800, 1)  # Limiting range for visualization
# Binomial probability mass functions
pmf_A = binom.pmf(k, nA, pA)
pmf_B = binom.pmf(k, nB, pB)

# Plot
plt.figure(figsize=(12, 6))
plt.plot(k, pmf_A, 'b-', label='Group A (Old Subject Line)')
plt.plot(k, pmf_B, 'r-', label='Group B (New Subject Line)')
plt.xlabel('Number of Clicks')
plt.ylabel('Probability')
plt.title('Binomial Distribution PMF for Clicks')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
num_samples = 1000
sample_size = 30

# Generate sample means
sample_means = [np.mean(np.random.randint(1, 7, sample_size)) for _ in range(num_samples)]

# Plot the distribution of sample means
# plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='b')
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import poisson

# Parameters
lambda_rate = 20  # Average rate of visits per hour

# Number of visits to calculate probabilities for
k_values = [15, 20, 25]

# Calculate and print probabilities
for k in k_values:
    probability = poisson.pmf(k, lambda_rate)
    print(f"Probability of exactly {k} visits: {probability:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define range of k (number of events)
k = np.arange(0, 21)  # Number of events from 0 to 20

# Define different lambda values
lambdas = [1,3, 5, 10, 20, 30]


for lam in lambdas:
    pmf = poisson.pmf(k, lam)
    plt.plot(k, pmf, marker='o', label=f' = {lam}')

plt.title('Poisson Distribution PMF for Different  Values')
plt.xlabel('Number of Events (k)')
plt.ylabel('Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Define parameters
lambda_rate = 1  # Rate of 10 calls per hour

# Define the range of time
x = np.linspace(0, 1, 100)

# Calculate the PDF
pdf = expon.pdf(x, scale=1/lambda_rate)

# Plot the PDF
plt.figure(figsize=(10, 6))

plt.plot(x, pdf, label=f' = {lambda_rate}')
plt.title('Exponential Distribution PDF')
plt.xlabel('Time (hours)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

# Calculate the average time between calls
mean_time = 1 / lambda_rate
print(f"Average time between calls: {mean_time:.2f} hours")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Parameters
lambda_rate = 1/5  # Average rate of 1 decay per 5 minutes

# Define range of time
x = np.linspace(0, 10, 100)

# Calculate the PDF
pdf = expon.pdf(x, scale=1/lambda_rate)

# Calculate the CDF
cdf = expon.cdf(x, scale=1/lambda_rate)



plt.subplot(1, 2, 1)
plt.plot(x, pdf, label=f' = {lambda_rate}')
plt.title('Exponential Distribution PDF')
plt.xlabel('Time (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Plot CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, label=f' = {lambda_rate}')
plt.title('Exponential Distribution CDF')
plt.xlabel('Time (minutes)')
plt.ylabel('Cumulative Probability')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Parameters for the Bernoulli distribution
p = 0.7  # Probability of success

# Define the possible outcomes (0 = Failure, 1 = Success)
x = [0, 1]
# Corresponding probabilities
pmf = bernoulli.pmf(x, p)

# Plot the Bernoulli distribution
plt.bar(x, pmf, color='blue', alpha=0.7, width=0.3)
plt.xticks([0, 1], ['Failure (0)', 'Success (1)'])
plt.title('Bernoulli Distribution (p=0.7)')
plt.ylabel('Probability')
plt.xlabel('Outcome')
plt.ylim(0, 1)
plt.show()
```
